```python
transactions['risk_score'] = (
    transactions['device_risk_score'] * 0.3 +
    transactions['merchant_risk_score'] * 0.2 +
    transactions['velocity_1h'] * 5 +
    transactions['new_device_flag'] * 15 +
    transactions['is_night_flag'] * 10 +
    transactions['geo_distance_km'] * 0.02
)
```

```python
transactions[['risk_score', 'fraud_label']].head()
```

```python
transactions.groupby('fraud_label')['risk_score'].mean()
```

```python
sns.boxplot(x='fraud_label', y='risk_score', data=transactions)

plt.title('Risk Score by Fraud Status')

plt.show()
```

```python
high_risk = transactions.sort_values(by='risk_score', ascending=False)

high_risk[['transaction_id', 'risk_score', 'fraud_label']].head(10)
```

```python
def risk_category(score):
    if score >= 80:
        return 'High Risk'
    elif score >= 50:
        return 'Medium Risk'
    else:
        return 'Low Risk'

transactions['risk_category'] = transactions['risk_score'].apply(risk_category)
```

```python
transactions['risk_category'].value_counts()
```

```python
transactions.groupby('risk_category')['fraud_label'].mean()
```

```python
risk_fraud = transactions.groupby('risk_category')['fraud_label'].mean()

risk_fraud.plot(kind='bar')

plt.title('Fraud Rate by Risk Category')

plt.xlabel('Risk Category')
plt.ylabel('Fraud Rate')

plt.show()
```
